```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader, TensorDataset
import time

# Start timing
start_time = time.time()

# Check if GPU is available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Prepare training data
data = np.array([[2/5, 17.12],
            ......
        [0.129982,2.69521]])

T_train = data[:, 0].reshape(-1, 1)  # Input data T
S_train = data[:, 1].reshape(-1, 1)  # Target output data S

# Convert to PyTorch tensors and move to GPU
T_train_tensor = torch.tensor(T_train, dtype=torch.float32).to(device)
S_train_tensor = torch.tensor(S_train, dtype=torch.float32).to(device)

# Create DataLoader
dataset = TensorDataset(T_train_tensor, S_train_tensor)
dataloader = DataLoader(dataset, batch_size=64, shuffle=True)

# Define the neural network model
class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(1, 16),
            nn.Sigmoid(),
            nn.Linear(16, 32),
            nn.Sigmoid(),
            nn.Linear(32, 16),
            nn.Sigmoid(),
            nn.Linear(16, 1)
        )
    
    def forward(self, x):
        return self.net(x)

model1= NeuralNetwork().to(device)

# Define loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model1.parameters(), lr=0.001)

# Training loop
num_epochs = 50000
losses = []

for epoch in range(num_epochs):
    epoch_start_time = time.time()
    
    for inputs, targets in dataloader:
        # Forward pass
        outputs = model1(inputs)
        loss = criterion(outputs, targets)
        
        # Backward pass and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    
    # Store loss for plotting
    with torch.no_grad():
        total_loss = criterion(model1(T_train_tensor), S_train_tensor)
        losses.append(total_loss.item())
    
    # Print progress
    if (epoch+1) % 5000 == 0:
        epoch_time = time.time() - epoch_start_time
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss.item():.4f}, Epoch Time: {epoch_time:.2f}s')

# Generate test data
T_try = np.linspace(0, 1, 200).reshape(-1, 1)
T_try_tensor = torch.tensor(T_try, dtype=torch.float32).to(device)

# Make predictions
with torch.no_grad():
    S_pred_tensor = model1(T_try_tensor)
    S_pred = S_pred_tensor.cpu().numpy()  # Move back to CPU for plotting

# Visualize results
plt.figure(figsize=(10, 6))
plt.scatter(T_train, S_train, color='blue', label='Actual Data')
plt.plot(T_try, S_pred, color='red', label='Model Prediction')
plt.xlabel('T')
plt.ylabel('S')
plt.title('Neural Network Regression')
plt.legend()
plt.grid(True)
plt.show()

# Plot training loss curve
plt.figure(figsize=(10, 6))
plt.plot(losses)
plt.title('Model Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.show()

# Calculate and print total runtime
total_time = time.time() - start_time
print(f"\nTotal runtime: {total_time:.2f} seconds")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# =============== A(z)  =================
class ConstrainedAModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(1, 16),
            nn.Tanh(),
            nn.Linear(16, 16),
            nn.Tanh(),
            nn.Linear(16, 16),
            nn.Tanh(),
            nn.Linear(16, 1),
            nn.Sigmoid()  # 输出在 (0,1)
        )
    
    def forward(self, inputs):
        ff = -self.net(inputs)       
        return ff * inputs           

# =============== f(z)  ========================
class ExpDecayF(nn.Module):
    def __init__(self, hidden=16, depth=4, act='tanh'):
        super().__init__()

        if act == 'tanh':
            Act = nn.Tanh
        elif act == 'silu':
            Act = nn.SiLU
        elif act == 'gelu':
            Act = nn.GELU
        else:
            raise ValueError("act must be one of: 'tanh', 'silu', 'gelu'")

        layers = []
        in_dim = 1
        for i in range(depth):
            layers += [nn.Linear(in_dim if i == 0 else hidden, hidden), Act()]
        layers += [nn.Linear(hidden, 1)]
        self.net = nn.Sequential(*layers)

    def forward(self, z):
        raw = self.net(z)      # raw ∈ (-∞, +∞)
        f = torch.exp(raw)    # f ∈ (0, +∞)
        return f


# ===============  ====================
def parallel_integrate_for_T(zh_batch, a_model, num_points=400):
    # x: (batch, num_points)
    x = torch.stack([
        torch.linspace(0.0, zh.item(), num_points, device=device)
        for zh in zh_batch
    ])  # shape: [B, N]

    x_flat = x.reshape(-1, 1)                     # [B*N, 1]
    A_flat = a_model(x_flat).squeeze()            # [B*N]
    A = A_flat.view_as(x)                         # [B, N]

    integrand = x**3 * torch.exp(-3.0 * A)        # [B, N]

    dx = zh_batch / (num_points - 1)              # [B]
    y = integrand
    integrals = (y[:, :-1] + y[:, 1:]).sum(dim=1) * dx / 2.0   # 

    return integrals   # [B]

def T(zh_batch, a_model, G):
 
    zh_batch = zh_batch.clone().detach()  # 
    integrals = parallel_integrate_for_T(zh_batch, a_model)
    nn_output = a_model(zh_batch.unsqueeze(-1)).squeeze()
    numerator = zh_batch**3 * torch.exp(-3.0 * nn_output)
    denominator = 4.0 * np.pi * integrals
    return numerator / denominator

def S(zh_batch, a_model, G):
    nn_output = a_model(zh_batch.unsqueeze(-1)).squeeze()
    return torch.exp(3.0 * nn_output) / (4.0 * G * zh_batch**3)

def parallel_integrate_for_rho(zh_batch, a_model, f_model, num_points=400):
    """
    I(z_h) = ∫_0^{z_h} y / (e^{A(y)} f(y)) dy
    用于 ρ/μ = 1 / [16πG * I(z_h)]
    """
    # y: (batch, num_points)
    y = torch.stack([
        torch.linspace(0.0, zh.item(), num_points, device=device)
        for zh in zh_batch
    ])  # [B, N]

    y_flat = y.reshape(-1, 1)                # [B*N, 1]
    A_flat = a_model(y_flat).squeeze()       # [B*N]
    f_flat = f_model(y_flat).squeeze()       # [B*N]  > 0
    A = A_flat.view_as(y)                    # [B, N]
    f = f_flat.view_as(y)                    # [B, N]

    f_safe = f.clamp_min(1e-6)

    integrand = y / (torch.exp(A) * f_safe)  # [B, N]

    dy = zh_batch / (num_points - 1)         # [B]
    I = (integrand[:, :-1] + integrand[:, 1:]).sum(dim=1) * dy / 2.0
    return I  # [B]

def Chi(zh_batch, a_model, f_model, G):
    I = parallel_integrate_for_rho(zh_batch, a_model, f_model)
    I_safe = I.clamp_min(1e-6)

    # T(z_h)
    T_values = T(zh_batch, a_model, G).clamp_min(1e-6)

    chi = 1.0 / (16.0 * np.pi * G * I_safe * T_values**2)

    return chi


def loss_fn(a_model, f_model, G, model1, model2):

    n_low, n_high, n_far = 30, 150, 30

    zh_low = torch.linspace(0.01, 0.6, n_low, device=device)
    zh_high = torch.linspace(0.6, 4.0, n_high, device=device)
    zh_far = torch.linspace(4.0, 20.0, n_far, device=device)

    zh_range = torch.cat([zh_low, zh_high, zh_far])    # [N]

    # T(z_h)
    T_values = T(zh_range, a_model, G)         # [N]

    S_pred = model1(T_values.unsqueeze(-1)).squeeze()
    Chi_pred = model2(T_values.unsqueeze(-1)).squeeze()

    S_values = S(zh_range, a_model, G) / T_values**3
    Chi_values = Chi(zh_range, a_model, f_model, G)

    loss_S = torch.mean((S_pred - S_values)**2)
    loss_Chi = torch.mean((Chi_pred - Chi_values)**2)

    return loss_S + 6000.0 * loss_Chi

# ============================================

a_model = ConstrainedAModel().to(device)
f_model = ExpDecayF().to(device)


PARAM_RANGES = {
    'G': (0.2, 0.8)
}

G_init = max(min(0.42, PARAM_RANGES['G'][1]), PARAM_RANGES['G'][0])
G_optimal = torch.tensor(G_init, requires_grad=True, device=device)


optimizer = optim.Adam([
    {'params': a_model.parameters()},
    {'params': f_model.parameters()},
    {'params': [G_optimal]}
], lr=1e-4)

scheduler = optim.lr_scheduler.ExponentialLR(optimizer, gamma=0.99)

best_loss = float('inf')
best_params = None
best_state = None   

def clamp_parameters():
    with torch.no_grad():
        G_optimal.copy_(G_optimal.clamp(*PARAM_RANGES['G']))


for epoch in range(50000):
    optimizer.zero_grad()
    loss = loss_fn(a_model, f_model, G_optimal, model1, model2)
    loss.backward()
    optimizer.step()

    clamp_parameters()

    if loss.item() < best_loss:
        best_loss = loss.item()
        best_state = {
            'a_model': {k: v.clone().detach() for k, v in a_model.state_dict().items()},
            'f_model': {k: v.clone().detach() for k, v in f_model.state_dict().items()}
        }
        best_params = {
            'G': G_optimal.item()
        }

    if epoch % 5000 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.6f}")
        print(f"Current G: {G_optimal.item():.6f}")
        scheduler.step()

if best_state is not None:
    a_model.load_state_dict(best_state['a_model'])
    f_model.load_state_dict(best_state['f_model'])
    print("Loaded best model state.")

final_G = max(min(best_params['G'], PARAM_RANGES['G'][1]), PARAM_RANGES['G'][0])

print("\nOptimal Parameters:")
print(f"G: {final_G:.6f}  (range: {PARAM_RANGES['G'][0]} to {PARAM_RANGES['G'][1]})")
print(f"\nMinimum Loss: {best_loss:.6f}")

torch.save({
    'a_model_state_dict': a_model.state_dict(),
    'f_model_state_dict': f_model.state_dict(),
    'params': best_params
}, 'best_model_A_f.pth')
print("Saved best model to 'best_model_A_f.pth'")
```

```python
import time
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim

# =========================================================
# device
# =========================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

model1.eval()
model2.eval()
for p in model1.parameters():
    p.requires_grad_(False)
for p in model2.parameters():
    p.requires_grad_(False)

data_rhoT3 = np.array([
    [0.1350182168309649,  0.04890965732087227],
    ......
    [0.1779058379208808,  0.2046728971962617],
], dtype=np.float64)

T_r_tab = torch.tensor(data_rhoT3[:, 0], dtype=torch.float32, device=device)
rhoT3_tab = torch.tensor(data_rhoT3[:, 1], dtype=torch.float32, device=device)

def rhoT3_target_from_T(T_query: torch.Tensor) -> torch.Tensor:
    # 线性插值：T_query -> rho/T^3
    Tq = T_query.clamp(T_r_tab.min(), T_r_tab.max())
    idx = torch.searchsorted(T_r_tab, Tq, right=True) - 1
    idx = idx.clamp(0, T_r_tab.numel() - 2)

    T0 = T_r_tab[idx]
    T1 = T_r_tab[idx + 1]
    y0 = rhoT3_tab[idx]
    y1 = rhoT3_tab[idx + 1]
    w = (Tq - T0) / (T1 - T0 + 1e-12)
    return y0 + w * (y1 - y0)

class ConstrainedAModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(1, 16),
            nn.Tanh(),
            nn.Linear(16, 16),
            nn.Tanh(),
            nn.Linear(16, 16),
            nn.Tanh(),
            nn.Linear(16, 1),
            nn.Sigmoid()  # (0,1)
        )

    def forward(self, z):
        ff = -self.net(z)      # (-1,0)
        return ff * z          # A(z)

class ExpDecayF(nn.Module):
    def __init__(self, hidden=16, depth=4, act='tanh'):
        super().__init__()
        if act == 'tanh':
            Act = nn.Tanh
        elif act == 'silu':
            Act = nn.SiLU
        elif act == 'gelu':
            Act = nn.GELU
        else:
            raise ValueError("act must be one of: 'tanh', 'silu', 'gelu'")

        layers = []
        for i in range(depth):
            layers += [nn.Linear(1 if i == 0 else hidden, hidden), Act()]
        layers += [nn.Linear(hidden, 1)]
        self.net = nn.Sequential(*layers)

    def forward(self, z):
        raw = self.net(z)
        return torch.exp(raw)  # (0,+∞)

def _cumtrapz(y: torch.Tensor, dz: torch.Tensor) -> torch.Tensor:
    """
    y: [B,N]
    dz: [B] or scalar
    return: [B,N]  with u[:,0]=0 and u[:,k]=∫_0^{z_k} y dz
    """
    if dz.dim() == 1:
        dz = dz.unsqueeze(1)  # [B,1]
    inc = 0.5 * (y[:, 1:] + y[:, :-1]) * dz
    out = torch.zeros_like(y)
    out[:, 1:] = torch.cumsum(inc, dim=1)
    return out

def integrals_I1_I2_I4(zh_batch, a_model, f_model, num_points=600):
    """
    I1 = ∫_0^{zh} z^3 e^{-3A} dz
    I2 = ∫_0^{zh} z / (e^A f) dz
    I4 = ∫_0^{zh} z^3 e^{-3A} (∫_0^z t/(e^A f) dt) dz
    """
    B = zh_batch.shape[0]
    t = torch.linspace(0.0, 1.0, num_points, device=zh_batch.device).unsqueeze(0)  # [1,N]
    z = zh_batch.unsqueeze(1) * t                                                  # [B,N]
    dz = zh_batch / (num_points - 1)                                                # [B]

    z_flat = z.reshape(-1, 1)
    A_flat = a_model(z_flat).squeeze(-1)
    f_flat = f_model(z_flat).squeeze(-1).clamp_min(1e-8)

    A = A_flat.view(B, num_points)
    f = f_flat.view(B, num_points)

    y1 = z**3 * torch.exp(-3.0 * A)
    u1 = _cumtrapz(y1, dz)

    y2 = z / (torch.exp(A) * f)
    u2 = _cumtrapz(y2, dz)

    y3 = z**3 * torch.exp(-3.0 * A) * u2
    u3 = _cumtrapz(y3, dz)

    I1 = u1[:, -1].clamp_min(1e-12)
    I2 = u2[:, -1].clamp_min(1e-12)
    I4 = u3[:, -1]  
    Azh = A[:, -1]
    return I1, I2, I4, Azh

_PI4_INV = 1.0 / (4.0 * np.pi)

def T_mu0_from_integrals(zh, I1, Azh):
    # T(mu=0) = (1/4π) * zh^3 e^{-3A(zh)} / I1
    return (_PI4_INV * zh**3 * torch.exp(-3.0 * Azh) / I1).clamp_min(1e-8)

def S_over_T3_mu0(zh, Azh, T0, G):
    # S/T^3 = [e^{3A}/(4G zh^3)] / T^3
    S = torch.exp(3.0 * Azh) / (4.0 * G * zh**3)
    return S / (T0**3)

def Chi_mu0_from_integrals(I2, T0, G):
    # chi = 1/(16π G I2 T^2)
    chi = 1.0 / (16.0 * np.pi * G * I2 * T0**2)
    return chi.clamp_min(1e-12)

def T_sym_from_integrals(zh, mu, I1, I2, I4, Azh):
    # T(zh,mu) = base * [1 - (mu/I2)^2 (I1*I2 - I4)]
    base = _PI4_INV * zh**3 * torch.exp(-3.0 * Azh) / I1
    bracket = 1.0 - (mu / I2)**2 * (I1 * I2 - I4)
    return (base * bracket).clamp_min(1e-8)

def solve_mu_eq_T_cached(zh, I1, I2, I4, Azh, n_iter=12):
    mu = T_mu0_from_integrals(zh, I1, Azh).clamp_min(1e-6)
    for _ in range(n_iter):
        Tmu = T_sym_from_integrals(zh, mu, I1, I2, I4, Azh)
        mu = (0.7 * mu + 0.3 * Tmu).clamp_min(1e-6)
    T_final = T_sym_from_integrals(zh, mu, I1, I2, I4, Azh).clamp_min(1e-6)
    return mu, T_final

def rho_over_T3_mu_eq_T_cached(zh, I1, I2, I4, Azh, G):
    mu_star, T_star = solve_mu_eq_T_cached(zh, I1, I2, I4, Azh)
    rho = mu_star / (16.0 * np.pi * G * I2)
    return (rho / T_star**3).clamp_min(1e-12), T_star

a_model = ConstrainedAModel().to(device)
f_model = ExpDecayF().to(device)

ckpt_path = "best_model_A_f.pth"
ckpt = torch.load(ckpt_path, map_location=device)

a_model.load_state_dict(ckpt["a_model_state_dict"])
f_model.load_state_dict(ckpt["f_model_state_dict"])
G_init = float(ckpt["params"]["G"])

G_opt = torch.tensor(G_init, device=device, requires_grad=True)
print(f"\nLoaded {ckpt_path}, initial G={G_init:.6f}")

optimizer = optim.Adam(
    [
        {"params": a_model.parameters()},
        {"params": f_model.parameters()},
        {"params": [G_opt]},
    ],
    lr=1e-4,
)

def clamp_G():
    with torch.no_grad():
        G_opt.clamp_(0.2, 0.8)

# =========================================================
# Range configs (each constraint has its own fit window)
# =========================================================
RANGE_CFG = {
    "S":   {"zh_min": 0.1, "zh_max": 5.0, "n_zh": 120, "T_min": 0.05, "T_max": 0.60},
    "Chi": {"zh_min": 0.1, "zh_max": 5.0, "n_zh": 120, "T_min": 0.05, "T_max": 0.60},
    "Rho": {"zh_min": 0.1, "zh_max": 5.0, "n_zh": 240, "T_min": 0.1350182168309649, "T_max": 0.1779058379208808},
}

def sample_zh(cfg):
    return torch.linspace(cfg["zh_min"], cfg["zh_max"], cfg["n_zh"], device=device)

def make_T_mask(T, cfg):
    Tmin = torch.tensor(cfg["T_min"], device=T.device, dtype=T.dtype)
    Tmax = torch.tensor(cfg["T_max"], device=T.device, dtype=T.dtype)
    return (T >= Tmin) & (T <= Tmax)

# =========================================================
# 6) 联合损失：S + Chi + rho/T^3(mu=T)
# =========================================================
def joint_loss(
    a_model,
    f_model,
    G,
    wS=10.0,
    wChi=1000.0,
    wRho=5000.0,
    num_points_mu0=600,
    num_points_muT=600,
    range_cfg=RANGE_CFG,
):
    # -------------------------
    # (A) mu=0 : S constraint
    # -------------------------
    cfgS = range_cfg["S"]
    zhS = sample_zh(cfgS)

    I1S, I2S, I4S, AzhS = integrals_I1_I2_I4(zhS, a_model, f_model, num_points=num_points_mu0)
    T0S = T_mu0_from_integrals(zhS, I1S, AzhS)

    S_bulk = S_over_T3_mu0(zhS, AzhS, T0S, G)
    S_tgt = model1(T0S.unsqueeze(-1)).squeeze(-1)

    maskS = make_T_mask(T0S, cfgS)
    loss_S = ((S_tgt[maskS] - S_bulk[maskS]) ** 2).mean() if maskS.any() else torch.zeros((), device=device)

    # -------------------------
    # (B) mu=0 : Chi constraint
    # -------------------------
    cfgC = range_cfg["Chi"]
    zhC = sample_zh(cfgC)

    I1C, I2C, I4C, AzhC = integrals_I1_I2_I4(zhC, a_model, f_model, num_points=num_points_mu0)
    T0C = T_mu0_from_integrals(zhC, I1C, AzhC)

    Chi_bulk = Chi_mu0_from_integrals(I2C, T0C, G)
    Chi_tgt = model2(T0C.unsqueeze(-1)).squeeze(-1)

    maskC = make_T_mask(T0C, cfgC)
    loss_Chi = ((Chi_tgt[maskC] - Chi_bulk[maskC]) ** 2).mean() if maskC.any() else torch.zeros((), device=device)

    # -------------------------
    # (C) mu=T : rho/T^3 constraint
    # -------------------------
    cfgR = range_cfg["Rho"]
    zhR = sample_zh(cfgR)

    I1R, I2R, I4R, AzhR = integrals_I1_I2_I4(zhR, a_model, f_model, num_points=num_points_muT)
    rhoT3_bulk, T_muT = rho_over_T3_mu_eq_T_cached(zhR, I1R, I2R, I4R, AzhR, G)

    maskR1 = make_T_mask(T_muT, cfgR)
    Tmin_data, Tmax_data = T_r_tab.min(), T_r_tab.max()
    maskR2 = (T_muT >= Tmin_data) & (T_muT <= Tmax_data)
    maskR = maskR1 & maskR2

    if maskR.any():
        rhoT3_tgt = rhoT3_target_from_T(T_muT[maskR])
        loss_Rho = ((rhoT3_bulk[maskR] - rhoT3_tgt) ** 2).mean()
    else:
        loss_Rho = torch.zeros((), device=device)

    total = wS * loss_S + wChi * loss_Chi + wRho * loss_Rho
    return total, loss_S.detach(), loss_Chi.detach(), loss_Rho.detach()

print("\n=== Fine-tuning A,f,G with (S,Chi,rho/T^3 @ mu=T) ===")
best = float("inf")
best_state = None

t0 = time.time()
for epoch in range(1000):
    optimizer.zero_grad(set_to_none=True)

    loss, lS, lC, lR = joint_loss(
        a_model, f_model, G_opt,
        wS=10.0,
        wChi=1000.0,
        wRho=5000.0,
        num_points_mu0=600,
        num_points_muT=600,
    )

    loss.backward()
    optimizer.step()
    clamp_G()

    cur = float(loss.detach().cpu().item())
    if cur < best:
        best = cur
        best_state = {
            "a": {k: v.detach().clone() for k, v in a_model.state_dict().items()},
            "f": {k: v.detach().clone() for k, v in f_model.state_dict().items()},
            "G": float(G_opt.detach().cpu().item()),
        }

    if epoch % 200 == 0:
        print(
            f"epoch {epoch:5d}  loss={cur:.3e} | "
            f"S={lS.item():.3e} Chi={lC.item():.3e} Rho={lR.item():.3e} | "
            f"G={G_opt.item():.6f}"
        )

print(f"\nFine-tune runtime: {time.time() - t0:.2f}s  best_loss={best:.6e}")

if best_state is not None:
    a_model.load_state_dict(best_state["a"])
    f_model.load_state_dict(best_state["f"])
    with torch.no_grad():
        G_opt.copy_(torch.tensor(best_state["G"], device=device))
    print(f"Loaded best fine-tuned state, G={G_opt.item():.6f}")

torch.save(
    {
        "a_model_state_dict": a_model.state_dict(),
        "f_model_state_dict": f_model.state_dict(),
        "params": {"G": float(G_opt.detach().cpu().item())},
        "note": "fine-tuned with S, Chi (mu=0) and rho/T^3 (mu=T) | unified integrals + cached mu-iteration",
    },
    "best_model_A_f_finetune_S_Chi_Rho.pth",
)
print("Saved: best_model_A_f_finetune_S_Chi_Rho.pth")
```

```python
import matplotlib.pyplot as plt
import torch

@torch.no_grad()
def plot_compare_S_Chi_Rho(
    a_model, f_model, G,
    model1, model2,
    RANGE_CFG, T_r_tab, rhoT3_target_from_T,
    num_points_mu0=800,
    num_points_muT=800
):
    a_model.eval()
    f_model.eval()

    # =========================
    # 1) S/T^3 (mu=0)
    # =========================
    cfgS = RANGE_CFG["S"]
    zhS = torch.linspace(cfgS["zh_min"], cfgS["zh_max"], cfgS["n_zh"], device=device)

    I1S, I2S, I4S, AzhS = integrals_I1_I2_I4(zhS, a_model, f_model, num_points=num_points_mu0)
    T0S = T_mu0_from_integrals(zhS, I1S, AzhS)

    S_bulk = S_over_T3_mu0(zhS, AzhS, T0S, G)              # bulk: S/T^3
    S_tgt  = model1(T0S.unsqueeze(-1)).squeeze(-1)          # target from model1

    maskS = make_T_mask(T0S, cfgS)

    Ts = T0S[maskS].detach().cpu().numpy()
    ys_bulk = S_bulk[maskS].detach().cpu().numpy()
    ys_tgt  = S_tgt[maskS].detach().cpu().numpy()

    order = Ts.argsort()
    Ts, ys_bulk, ys_tgt = Ts[order], ys_bulk[order], ys_tgt[order]

    plt.figure()
    plt.plot(Ts, ys_tgt, label="Target (model1)")
    plt.plot(Ts, ys_bulk, label="Bulk (NN)")
    plt.xlabel("T")
    plt.ylabel("S/T^3")
    plt.title("S/T^3 (mu=0): Bulk vs Target")
    plt.legend()
    plt.tight_layout()


    # =========================
    # 2) Chi (mu=0)
    # =========================
    cfgC = RANGE_CFG["Chi"]
    zhC = torch.linspace(cfgC["zh_min"], cfgC["zh_max"], cfgC["n_zh"], device=device)

    I1C, I2C, I4C, AzhC = integrals_I1_I2_I4(zhC, a_model, f_model, num_points=num_points_mu0)
    T0C = T_mu0_from_integrals(zhC, I1C, AzhC)

    Chi_bulk = Chi_mu0_from_integrals(I2C, T0C, G)
    Chi_tgt  = model2(T0C.unsqueeze(-1)).squeeze(-1)

    maskC = make_T_mask(T0C, cfgC)

    Tc = T0C[maskC].detach().cpu().numpy()
    yc_bulk = Chi_bulk[maskC].detach().cpu().numpy()
    yc_tgt  = Chi_tgt[maskC].detach().cpu().numpy()

    order = Tc.argsort()
    Tc, yc_bulk, yc_tgt = Tc[order], yc_bulk[order], yc_tgt[order]

    plt.figure()
    plt.plot(Tc, yc_tgt, label="Target (model2)")
    plt.plot(Tc, yc_bulk, label="Bulk (NN)")
    plt.xlabel("T")
    plt.ylabel("Chi")
    plt.title("Chi (mu=0): Bulk vs Target")
    plt.legend()
    plt.tight_layout()


    # =========================
    # 3) rho/T^3 (mu=T)
    # =========================
    cfgR = RANGE_CFG["Rho"]
    zhR = torch.linspace(cfgR["zh_min"], cfgR["zh_max"], cfgR["n_zh"], device=device)

    I1R, I2R, I4R, AzhR = integrals_I1_I2_I4(zhR, a_model, f_model, num_points=num_points_muT)
    rhoT3_bulk, T_muT = rho_over_T3_mu_eq_T_cached(zhR, I1R, I2R, I4R, AzhR, G)

    maskR1 = make_T_mask(T_muT, cfgR)
    Tmin_data, Tmax_data = T_r_tab.min(), T_r_tab.max()
    maskR2 = (T_muT >= Tmin_data) & (T_muT <= Tmax_data)
    maskR = maskR1 & maskR2

    Tr = T_muT[maskR]
    yr_bulk = rhoT3_bulk[maskR]
    yr_tgt  = rhoT3_target_from_T(Tr)

    Tr = Tr.detach().cpu().numpy()
    yr_bulk = yr_bulk.detach().cpu().numpy()
    yr_tgt  = yr_tgt.detach().cpu().numpy()

    order = Tr.argsort()
    Tr, yr_bulk, yr_tgt = Tr[order], yr_bulk[order], yr_tgt[order]

    plt.figure()
    plt.plot(Tr, yr_tgt, label="Target (table interp)")
    plt.plot(Tr, yr_bulk, label="Bulk (NN)")
    plt.xlabel("T (mu=T)")
    plt.ylabel("rho/T^3")
    plt.title("rho/T^3 (mu=T): Bulk vs Target")
    plt.legend()
    plt.tight_layout()

    plt.show()


plot_compare_S_Chi_Rho(
    a_model=a_model,
    f_model=f_model,
    G=G_opt,
    model1=model1,
    model2=model2,
    RANGE_CFG=RANGE_CFG,
    T_r_tab=T_r_tab,
    rhoT3_target_from_T=rhoT3_target_from_T,
    num_points_mu0=800,
    num_points_muT=800
)
```

```python
import torch
import numpy as np

@torch.no_grad()
def T_mu_picture(zh_batch, mu, a_model, f_model, num_points=1200, z0=0.0):
    Ts = []
    for zh in zh_batch:
        zh_val = zh.item()
        if zh_val <= z0:
            Ts.append(torch.tensor(float('nan'), device=device))
            continue

        y = torch.linspace(z0, zh_val, num_points, device=device)
        dy = (zh_val - z0) / (num_points - 1)

        y_in = y.view(-1, 1)
        A = a_model(y_in).view(-1)
        f = f_model(y_in).view(-1).clamp_min(1e-12)

        # I1 integrand
        g1 = y**3 * torch.exp(-3.0 * A)

        # I2 integrand
        g2 = y / (torch.exp(A) * f)

        # I1
        I1 = torch.trapz(g1, y).clamp_min(1e-12)

        seg2 = 0.5 * (g2[:-1] + g2[1:]) * dy
        I2_cum = torch.zeros_like(g2)
        I2_cum[1:] = torch.cumsum(seg2, dim=0)
        I2 = I2_cum[-1].clamp_min(1e-12)

        # I4 = ∫ g1(y) * I2(y) dy
        I4 = torch.trapz(g1 * I2_cum, y).clamp_min(1e-12)

        A_h = a_model(zh.view(1, 1)).view(())
        pref = (zh**3 * torch.exp(-3.0 * A_h)) / (4.0 * np.pi * I1)

        bracket = 1.0 - (mu / I2)**2 * (I1 * I2 - I4)
        Ts.append(pref * bracket)

    return torch.stack(Ts)


mu = torch.tensor(0.9, device=device)
zh = torch.linspace(3, 3.8, 400, device=device)

with torch.no_grad():
    T_vals = T_mu_picture(zh, mu, a_model, f_model, z0=0.0)  

mask = torch.isfinite(T_vals) & (T_vals > 0)
plt.figure(figsize=(7,5))
plt.plot(zh[mask].cpu().numpy(), T_vals[mask].cpu().numpy(), linewidth=2)
plt.xlabel(r"$z_h$")
plt.ylabel(r"$T(z_h)$")
plt.title(rf"$T(z_h)$ at $\mu={mu.item():.1f}$")
plt.grid(True)
plt.tight_layout()
plt.show()
```
